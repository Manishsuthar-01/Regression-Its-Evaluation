**1. What is Simple Linear Regression?**
>**Definition:**

>Simple Linear Regression is a statistical method used to model the relationship between two variables: one independent variable (predictor or X) and one dependent variable (response or Y). The goal is to find the best-fitting straight line (also called the regression line) that predicts the value of the dependent variable based on the independent variable.

>**Equation of Simple Linear Regression:**

>>**Y=a+bX+ε**

>Where:


>Y = Dependent variable (response)


>X = Independent variable (predictor)


>a = Intercept of the line (value of Y when X = 0)


>b = Slope of the line (change in Y for a one-unit change in X)


>ε = Error term (difference between observed and predicted Y)

>**Objective:**

>The main objective of SLR is to:

>>Predict the value of Y based on X.

>>Understand the strength and direction of the linear relationship between the variables.

>**Assumptions of Simple Linear Regression:**

>>**Linearity:** The relationship between X and Y is linear.

>>**Independence:** Observations are independent.

>>**Homoscedasticity:** Constant variance of residuals.

>>**Normality of errors:** The residuals (errors) are normally distributed.

>**Steps Involved:**

>>Collect data on the two variables.

>>Plot a scatter diagram to visualize the relationship.

>>Calculate the regression coefficients (a and b) using the least squares method:






$$
b = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sum (X_i - \bar{X})^2}
$$



$$
a = \bar{Y} - b\bar{X}
$$


>>Fit the regression line to the data.

>>Interpret results using statistical measures (e.g., R², p-values).

>**Example:**

>>Suppose we want to study the relationship between hours studied (X) and marks obtained (Y). If the regression equation is:


>>Y=50+5X

>This means:

>>The student who studies 0 hours is expected to score 50 marks.

>>For each additional hour of study, marks increase by 5.

>**Applications:**

>>**Business:** Forecasting sales based on advertising spend.

>>**Economics:** Predicting consumption based on income.

>>**Education:** Predicting student performance from study hours.

>>**Healthcare:** Estimating health outcomes based on lifestyle factors.

>**Advantages:**

>>Easy to implement and interpret.

>>Helps in prediction and forecasting.

>>Useful in hypothesis testing and decision-making.

>**Limitations:**

>>Only works for linear relationships.

>>Sensitive to outliers.

>>Assumes all assumptions are satisfied.

**2. What are the key assumptions of Simple Linear Regression?**
>**Linearity**

>>**Assumption:** There is a linear relationship between the independent variable X and Y the dependent variable

>>**Explanation:** The change in Y is proportional to the change in X. In a scatter plot, the data points should follow a straight-line pattern.

>>**Violation Example:** If the relationship is curved or exponential, linear regression is not appropriate.

>**Independence of Errors**

>>**Assumption:** The residuals (errors) are independent of each other.

>>**Explanation:** The value of one observation’s error should not influence another’s.

>>**Violation Example:** This is often a problem in time series data where values are autocorrelated (e.g., today's value depends on yesterday's).

>**Homoscedasticity (Constant Variance of Errors)**

>>**Assumption:** The variance of the residuals is constant across all levels of the independent variable.

>>**Explanation:** The spread of residuals should be roughly the same for all predicted values.

>>**Violation Example:** If residuals fan out or narrow as X increases (heteroscedasticity), predictions become unreliable.

>**Normality of Errors**

>>**Assumption:** The residuals are normally distributed.

>>**Explanation:** This is especially important for small sample sizes, as it affects confidence intervals and hypothesis tests.

>>**Violation Example:** Skewed or heavily tailed residual distributions can invalidate statistical inferences.

>**No or Minimal Multicollinearity**

>>**Note:** This applies to multiple linear regression, but in simple linear regression, since there's only one predictor, this is not relevant.

**3. What is heteroscedasticity, and why is it important to address in regression models?**
>Heteroscedasticity refers to a condition in regression analysis where the variance of the residuals (errors) is not constant across all levels of the independent variable(s). In simpler terms, the spread or “noise” in the data increases or decreases as the value of the independent variable changes.

>**Visual Explanation:**

>>In a scatter plot of residuals vs. predicted values:

>>**Homoscedasticity (ideal):** Residuals are evenly spread — forming a “cloud” shape.

>>**Heteroscedasticity:** Residuals fan out (like a cone) or narrow, showing increasing or decreasing variance.

>**Example:**

>>Suppose you're modeling income (Y) based on years of education (X). If people with higher education levels show a wider range of incomes, this could lead to heteroscedastic errors, because the variance in income isn't constant across all education levels.

>**Why Heteroscedasticity Is a Problem:**

>>**Biased Standard Errors:**

>>>It leads to inaccurate standard errors of the regression coefficients.

>>>As a result, t-tests and p-values may be invalid.

>>>This can cause incorrect conclusions about which variables are significant.

>>**Unreliable Confidence Intervals and Hypothesis Tests:**

>>>Confidence intervals may be too narrow or too wide.

>>>This affects the reliability of predictions and statistical inference.

>>**Efficiency Loss:**

>>>Ordinary Least Squares (OLS) estimators are still unbiased but not efficient (i.e., not the best estimators in terms of minimum variance).

>How to Detect Heteroscedasticity:

>>Plot residuals vs. fitted values (check for patterns or fanning).

>>Statistical tests like:

>>>Breusch-Pagan Test

>>>White Test

>>>Goldfeld-Quandt Test

>**How to Fix or Address It:**

>>Transform the dependent variable (e.g., log transformation).

>>Use Weighted Least Squares (WLS) instead of OLS.

>>Use robust standard errors (e.g., heteroscedasticity-consistent errors).

**4. What is Multiple Linear Regression?**
>**Definition:**

>Multiple Linear Regression (MLR) is a statistical technique used to model the relationship between one dependent variable (Y) and two or more independent variables (X₁, X₂, ..., Xₙ). It is an extension of Simple Linear Regression, which uses only one independent variable.

>**Regression Equation:**

##Y = a + b1X1 + b2X2 + ... + bnXn + ε

>Where:


>>Y = Dependent variable

>>X1,X2,....,Xn = Independent (predictor) variables


>>a = Intercept (value of Y when all X's = 0)

>>b1,b2,....,bn = Coefficients (slopes)

>>ε = Error term (residual)



>Each coefficient bi represents the effect of that independent variable on Y, holding all other variables constant.

>**Purpose of Multiple Linear Regression:**

>>To predict the value of a dependent variable using multiple predictors.

>>To understand how each predictor influences the outcome.

>>To control for other variables when analyzing the effect of one variable.

>**Example:**

>>Suppose you're predicting a student's final exam score (Y) based on:

>>Hours studied (X₁),

>>Attendance rate (X₂), and

>>Number of assignments submitted (X₃).

>>Then the model might look like:

>>>**Score = a + b1(Hours Studied) + b2(Attendance) + b3(Assignments) + ε**

>**Assumptions of Multiple Linear Regression:**

>>**Linearity:** The relationship between dependent and independent variables is linear.

>>**Independence of Errors:** Observations are independent.

>>**Homoscedasticity:** Constant variance of residuals.

>>**Normality of Residuals:** Residuals are normally distributed.

>>**No Multicollinearity:** Independent variables should not be highly correlated with each other.

>**Advantages:**

>>Can handle multiple predictors, improving model accuracy.

>>Helps in controlling for confounding variables.

>>Useful for forecasting and optimization.

>**Limitations:**

>>Requires large datasets for reliable results.

>>Sensitive to multicollinearity and outliers.

>>Interpretation becomes complex as more variables are added.

**6. Implement a Python program to fit a Simple Linear Regression model to
the following sample data:**

**● X = [1, 2, 3, 4, 5]**

**● Y = [2.1, 4.3, 6.1, 7.9, 10.2]**

**Plot the regression line over the data points.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape to 2D array
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predict values using the model
Y_pred = model.predict(X)

# Print model parameters
print(f"Intercept: {model.intercept_:.2f}")
print(f"Slope: {model.coef_[0]:.2f}")

# Plot original data points
plt.scatter(X, Y, color='blue', label='Actual Data')

# Plot regression line
plt.plot(X, Y_pred, color='red', label='Regression Line')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

**7. Fit a Multiple Linear Regression model on this sample data:**

**● Area = [1200, 1500, 1800, 2000]**

**● Rooms = [2, 3, 3, 4]**

**● Price = [250000, 300000, 320000, 370000]**

**Check for multicollinearity using VIF and report the results.**


```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.linear_model import LinearRegression

# Input data
data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

# Separate features and target
X = data[['Area', 'Rooms']]
y = data['Price']

# Fit the regression model
model = LinearRegression()
model.fit(X, y)

# Print model coefficients
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficients: Area = {model.coef_[0]:.2f}, Rooms = {model.coef_[1]:.2f}")

# Add constant term for VIF calculation
X_with_const = sm.add_constant(X)

# Calculate VIF for each predictor
vif_data = pd.DataFrame()
vif_data["Feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i)
                   for i in range(X_with_const.shape[1])]

# Display VIF results
print("\nVariance Inflation Factors (VIF):")
print(vif_data)
```

>**Regression Equation:**

>>>**Price = 103157.89 + 63.16 * Area + 34736.84 * Rooms**

>**Interpretation of Coefficients:**

>>**Intercept (103157.89):**

>>The estimated base price when both Area and Rooms are zero. (Not practically meaningful but mathematically required.)

>>**Area Coefficient (63.16):**

>>For every additional square foot of area, the price increases by approximately $63.16, assuming the number of rooms is constant.

>>**Rooms Coefficient (34736.84):**

>>Each additional room increases the price by approximately $34,736.84, assuming area is constant.

>**VIF Analysis (Multicollinearity Check)**

>Feature	VIF

>>**Area	7.74**

>>**Rooms	7.74**


>VIF values between 5 and 10 indicate moderate multicollinearity.

>**Since both predictors have a VIF ≈ 7.74, there is multicollinearity, but it’s not extreme.**



**8. : Implement polynomial regression on the following data:**

**● X = [1, 2, 3, 4, 5]**

**● Y = [2.2, 4.8, 7.5, 11.2, 14.7]**

**Fit a 2nd-degree polynomial and plot the resulting curve.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform input for polynomial regression (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
Y_pred = model.predict(X_range_poly)

# Print the coefficients
print("Model Coefficients:")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficient for x: {model.coef_[1]:.2f}")
print(f"Coefficient for x^2: {model.coef_[2]:.2f}")

# Plot original data points
plt.scatter(X, Y, color='blue', label='Actual Data')

# Plot polynomial regression curve
plt.plot(X_range, Y_pred, color='red', label='Polynomial Fit (Degree 2)')

# Labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2nd-Degree Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()

```

**9. Create a residuals plot for a regression model trained on this data:**

**● X = [10, 20, 30, 40, 50]**

**● Y = [15, 35, 40, 50, 65]**

**Assess heteroscedasticity by examining the spread of residuals.**


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Data preparation
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Step 2: Fit linear regression model
model = LinearRegression()
model.fit(X, Y)

# Step 3: Predict Y values and calculate residuals
Y_pred = model.predict(X)
residuals = Y - Y_pred

# Step 4: Plot residuals
plt.scatter(X, residuals, color='blue', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--', linewidth=1)
plt.title('Residuals Plot')
plt.xlabel('X (Input Variable)')
plt.ylabel('Residuals (Actual - Predicted)')
plt.grid(True)
plt.legend()
plt.show()

# Optional: Print model for reference
print(f"Model: y = {model.intercept_:.2f} + {model.coef_[0]:.2f}x")
```

>**This means:**

>>For every 1-unit increase in x, y increases by **1.15** units on average.

>>The intercept **6.50** is the estimated value of y when x=0.

>>The residuals don’t seem to follow a clear increasing or decreasing pattern.

>>They alternate signs and don’t increase in variance drastically.

>>So visually, the residuals are fairly evenly spread — this suggests:

>>No strong evidence of heteroscedasticity (good).

**10. : Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.**
>**Step 1: Addressing Multicollinearity**

>**What it is:**

>>Multicollinearity occurs when predictor variables are highly correlated, inflating variance of coefficient estimates and making interpretation unstable.

>**How to detect:**

>>Use Variance Inflation Factor (VIF) or correlation matrix.

>>High VIF (>5 or 10) indicates multicollinearity.

>**Solutions:**

>>**Feature Selection / Removal:**

>>>Remove or combine highly correlated variables.

>>>Example: If area and number of rooms are correlated, consider keeping one or combining them into a new feature like “area per room”.

>>**Dimensionality Reduction:**

>>>Use Principal Component Analysis (PCA) or Partial Least Squares (PLS) to create uncorrelated components.

>>>Reduces correlated features to orthogonal (uncorrelated) factors.

>>**Regularization:**

>>>Apply Ridge Regression (L2 penalty) or Lasso Regression (L1 penalty).

>>>These techniques shrink coefficients of correlated variables and reduce variance.

>**Step 2: Addressing Heteroscedasticity**

>**What it is:**

>>Heteroscedasticity means the variance of residuals changes across levels of predictors, violating OLS assumptions and affecting standard errors and confidence intervals.

>**How to detect:**

>>Plot residuals vs. predicted values.

>>Formal tests like Breusch-Pagan or White’s test.

>**Solutions:**

>>**Transform Target Variable:**

>>>Apply transformations like log, square root, or Box-Cox on house prices to stabilize variance.

>>>Example: Use log(Price) instead of raw Price.

>>**Weighted Least Squares (WLS):**

>>>Assign weights inversely proportional to variance of residuals to reduce influence of points with high variance.

>>**Use Robust Standard Errors:**

>>>Adjust standard errors using heteroscedasticity-consistent estimators (e.g., Huber-White).

>>**Model More Complex Relationships:**

>>>Sometimes heteroscedasticity arises from omitted variables or nonlinear relationships.

>>>Include interaction terms, polynomial terms, or use models like Random Forests that don't assume homoscedasticity.

>**Step 3: Model Validation and Diagnostics**

>>Split data into train/test or use cross-validation to check model generalizability.

>>Examine residual plots post-fixes to ensure heteroscedasticity and multicollinearity are mitigated.

>>Evaluate metrics like RMSE, MAE, and R².
